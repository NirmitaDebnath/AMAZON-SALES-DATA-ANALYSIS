# Exploratory Data Analysis

**In the ratings column there are 175794 null values**

**Replacing these null values with 0 as there may be no ratings so they are null**

```python
df[['ratings']]=df[['ratings']].fillna(0)
```

```python
df.isnull().sum()
```

**Using bfill to fill the null values of discount price with the actual price, so we can consider the discountprice = actualprice**

```python
df=df.bfill(axis ='columns')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

**Using bfill to fill the null values of discount price with actual price and dropped the remaing null values**

```python
df[['discount_price','actual_price']] = df[['discount_price','actual_price']].replace({'₹': '',
                                                                                      ',': ''}, regex=True)
```

```python
df[['discount_price','actual_price']] 
```

**Removed the ₹ and , sign make the analysis easy**

**Changing the datatype of ratings, no. of ratings, discounted price and actual price**

```python
df[['discount_price','actual_price']] = df[['discount_price','actual_price']].astype(str).astype(float)
```

```python
df[['discount_price','actual_price']] 
```

```python
df.info()
```

```python
df[['ratings']].value_counts()
```

```python
df[['ratings']] = df[['ratings']].astype(str).astype(float)
```

# Categories by Sales

```python
ds=df.groupby(['main_category'],as_index=False)['discount_price'].sum()
plt.figure(figsize=(8,6))
ax=sns.barplot(y='main_category',x='discount_price',data=ds,palette="dark")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='TOTAL SALES')
plt.title("TOTAL SALES OF EACH CATEGORY")
```

# Top 5 categories and sub-categories with maximum sales

```python
ds=df.groupby(['main_category','sub_category'],as_index=False)['discount_price'].sum()
ds
```

```python
ds=df.groupby(['main_category','sub_category'],as_index=False)['discount_price'].sum()
ds=ds.nlargest(5, 'discount_price', keep='all')
ds
```

```python
ds=df.groupby(['main_category'],as_index=False)['discount_price'].sum()
ds=ds.nlargest(5, 'discount_price', keep='all')
plt.figure(figsize=(8,3))
ax=sns.barplot(y='main_category',x='discount_price',data=ds,palette="dark")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='TOTAL SALES')
plt.title("TOP 5 MAXIMUM SOLD PRODUCT CATEGORY")
```

```python
ds=df.groupby(['sub_category'],as_index=False)['discount_price'].sum()
ds=ds.nlargest(5, 'discount_price', keep='all')
plt.figure(figsize=(8,3))
ax=sns.barplot(y='sub_category',x='discount_price',data=ds,palette="rocket")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='TOTAL SALES')
plt.title("TOP 5 MAXIMUM SOLD PRODUCTS")
```

# Categories by Average ratings

```python
dr=df.groupby(['main_category'],as_index=False)['ratings'].mean()
plt.figure(figsize=(8,9))
ax=sns.barplot(y='main_category',x='ratings',data=dr,palette="cubehelix")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='AVERAGE RATINGS')
plt.title("CATEGORIES AND AVERAGE RATINGS")
```

```python
dr=dr.nlargest(5,'ratings', keep='all')
plt.figure(figsize=(6,3))
ax=sns.barplot(y='main_category',x='ratings',data=dr,palette="cubehelix")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='AVERAGE RATINGS')
plt.title("TOP 5 CATEGORIES WITH MAXIMUN AVERAGE RATINGS")
```

```python
dr=df.groupby(['sub_category'],as_index=False)['ratings'].mean()
dr=dr.nlargest(5,'ratings', keep='all')
plt.figure(figsize=(8,3))
ax=sns.barplot(y='sub_category',x='ratings',data=dr,palette="dark")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='AVERAGE RATINGS')
plt.title("TOP 5 Products with maximun Average Ratings")
```

# Categories by Number of ratings

```python
dr=df.groupby(['main_category'],as_index=False)['ratings'].count()
plt.figure(figsize=(8,9))
ax=sns.barplot(y='main_category',x='ratings',data=dr,palette="magma")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='NUMBER OF RATINGS GIVEN')
plt.title("NUMBER OF RATINGS GIVEN TO EACH CATEGORY")
```

```python
dr=df.groupby(['main_category'],as_index=False)['ratings'].count()
dr=dr.nlargest(10,'ratings', keep='all')
plt.figure(figsize=(8,5))
ax=sns.barplot(y='main_category',x='ratings',data=dr,palette="rocket")
for bars in ax.containers:ax.bar_label(bars) 
ax.set(ylabel='CATEGORIES', xlabel='NUMBER OF RATINGS GIVEN')
plt.title("TOP 10 CATEGORIES THAT RECEIVED MAXIMUM NUMBER OF RATINGS")
```
